```python
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id =product_id
        self.name =name
        self.price= float(price)
        self.quantity = int(quantity)

    def sell(self, qty):
        if qty<=self.quantity:
            self.quantity -=qty
            total = qty * self.price
            print("Order Processed Successfully!\n")
            print(f"Total Bill Amount = ${total:.2f}")
        else:
            print("Insufficient Stock!\n")
        
    def __str__(self):
        return f"{self.name} (ID: {self.product_id}) | Stock: {self.quantity} | Price: ${self.price:.2f}"

class InventorySystem:
    def __init__(self):
        self.inventory = {}

    def display_inventory(self):
        print("--- INVENTORY LIST ---\n")
        if not self.inventory:
            print("Inventory is empty.")
        else:
            for product in self.inventory.values():
                print(product)
    
    
    def add_or_update_product(self, pid, name, price, qty):
        if pid in self.inventory:
            self.inventory[pid].quantity += qty
            print(f"Updated {name} quantity.")
        else:
            self.inventory[pid] = Product(pid, name, price, qty)
            print(f"Added new product: {name}.")
            

    
    def process_sale(self):
        pid = input("Enter Product ID: \n")

        if pid in self.inventory:
            qty = int(input("Enter Quantity Sold: "))
            self.inventory[pid].sell(qty)
        else:
            print("Product Not Found!\n")

if __name__ == "__main__":

    
    system = InventorySystem()
    while True:
        
        print("\n==============================================")
        print(" SUPPLY CHAIN & INVENTORY MANAGEMENT SYSTEM")
        print("==============================================")
        print("1. View Inventory Dashboard")
        print("2. Inbound Logistics (Restock/Add Product)")
        print("3. Outbound Logistics (Process Order Sale)")
        print("4. Exit")
        print("==============================================")

        choice = input("Enter Your Choice: ")
        if choice == "1":
            system.display_inventory()
        elif choice == "2":
            p_id = input("Enter Product ID")
            p_name = input("Enter Product Name")
            p_qty = input("Enter Product Quantity")
            p_price = input("Enter Product price")
            system.add_or_update_product(p_id, p_name, p_price, p_qty)
          
        elif choice == "3":
            system.process_sale()

        elif choice == "4":
            print("\nThank You for Using the Inventory Management System!")
            break

        else:
            print("\nInvalid Choice! Please Try Again.")



    # # p_id = input("Enter Product ID")
    # # p_name = input("Enter Product Name")
    
    # # p_price = input("Enter Product price")
    # # p_qty = input("Enter Product Quantity")
    
    # system.add_or_update_product("101", "Wireless Mouse", 25.00, 12)

    # system.add_or_update_product("102", "HDMI Cable", 10.50, 3)

    # system.add_or_update_product("103", "Gaming Keyboard", 80.00, 2)

    # # system.add_or_update_product(p_id, p_name, p_qty, p_price)

    
    # system.display_inventory()
    # system.process_sale()
```
