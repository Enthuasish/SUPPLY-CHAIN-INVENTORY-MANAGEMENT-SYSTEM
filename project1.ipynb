```python
# Project 1: SUPPLY CHAIN & INVENTORY MANAGEMENT SYSTEM
# 1.	View Inventory Dashboard
# 2.	Inbound Logistics: Restock/Add Product
# 3.	Outbound Logistics: Process Order Sale
# 4.	Exit System

class Product:
    def __init__(self, pid, pname, price, quantity):
        self.pid = pid
        self.pname = pname
        self.price = float(price)
        self.quantity = int(quantity)

    def __str__(self):
        return f"{self.pname} : {self.pid} | {self.price} | {self.quantity}"
    
    def restock(self, qty):
        self.quantity += qty

    def sell(self, qty):
        if qty <= self.quantity:
            self.quantity -= qty
            return True
        else:
            return False

    
class SystemManager:
    def __init__(self):
        self.inventory = {}
    
    def add_or_restock(self, p_id, p_name, p_price, p_qty):
        if p_id in self.inventory:
            # self.inventory[p_id].quantity += p_qty
            self.inventory[p_id].restock(p_qty)
            print("Updated successfully")
        else:
            self.inventory[p_id] = Product( p_id, p_name, p_price, p_qty)
            print(f"{p_name} added successfully.")
    
    def ProcessSellOrder(self):
        pid = input("Enter Product id").upper()
        if pid in self.inventory:
    
            product = self.inventory[pid]
            q = int(input("Enter quantity"))
            if product.sell(q):

                print("---------------- Total Bill ----------------\n")
                print(f"Product ID: {product.pid}")
                print(f"Product name: {product.pname}")
                print(f"Price per unit: {product.price}")
                print(f"Quantity purchased : {q}")
                print(f"Total Amount: {product.price * q}")
                
                # self.inventory[pid].sell(q)
                print("Order processed Successfully\n")
            else:
                print("Insufficient Stock")
        else:
            print("Product is not available at this momemnt")

    def view_inventory(self):
        print("--- INVENTORY LIST ---\n")
        if not self.inventory:
            print("Inventory is Empty")
        else:
            for product in self.inventory.values():
                print(product)

def main():
    manager = SystemManager()

    while True:
        print("\n==============================================")
        print(" SUPPLY CHAIN & INVENTORY MANAGEMENT SYSTEM")
        print("==============================================")
        print("1. View Inventory Dashboard")
        print("2. Inbound Logistics (Restock/Add Product)")
        print("3. Outbound Logistics (Process Order Sale)")
        print("4. Exit")
        print("==============================================")

        choice = int(input("Enter Choice: "))

        if choice == 1:
            manager.view_inventory()
        elif choice == 2:

            pid = input("Enter the Product ID")
            if pid not in manager.inventory:
                pname = input("Enter the Product Name")
                price = input("Enter the Product Price")
                qnty = int(input("Enter the Product Quantity"))
                manager.add_or_restock(pid, pname, price, qnty)
            else:
                
                qnty = int(input("Enter the Product Quantity"))
                manager.add_or_restock(pid, pname, price, qnty)
        elif choice == 3:
            manager.ProcessSellOrder()
        elif choice == 4:
            print("\nThank You for Using the Inventory Management System!")
            break
        else:
            price("Invalid Choice ! Try Again")

if __name__ == "__main__":
    main()

    

        
        
```
